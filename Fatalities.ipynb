```python
import pandas as pd
import requests
import json
from pprint import pprint
from datetime import datetime, timedelta
from api_keys import api_key
```

```python
crashes_data = pd.read_csv('Crashes_in_DC.csv', low_memory=False)
crashes_data.head()
```

```python
crash_details = pd.read_csv('Crash_Details_Table.csv', low_memory=False)
crash_details.head()
```

```python
merged_data = crashes_data.merge(crash_details, on='CRIMEID', how='left')
merged_data.columns
```

```python
merged_data['WARD'].value_counts()
```

```python
non_dc_fatalities = merged_data[(merged_data['WARD']=='Null') & (merged_data['FATAL']=='Y') ]
non_dc_fatalities
```

```python
fatalities_filter = merged_data[merged_data['FATAL']=='Y']
fatalities_filter.head()

fatalities = fatalities_filter[[
    'REPORTDATE','AGE', 'WARD', 'FATAL_BICYCLIST', 'FATAL_DRIVER', 'FATAL_PEDESTRIAN', 'IMPAIRED', 'TOTAL_VEHICLES',
    'TOTAL_BICYCLES', 'TOTAL_PEDESTRIANS', 'PEDESTRIANSIMPAIRED', 'BICYCLISTSIMPAIRED', 'DRIVERSIMPAIRED',
    'LATITUDE', 'LONGITUDE']]

fatalities['WARD'].value_counts()
```

```python
fatalities.head()
```

```python
sunrise_list = []
sunset_list = []
weather_list = []
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
unit = "imperial"

# Work with 4 rows of data.
subset = fatalities[fatalities['AGE'] == 37]

# Determine weather / sunset related to crashes
for row, items in subset.iterrows():
    lat = items[13]
    lon = items[14]
    
    query_url = f"{url}APPID={api_key}&units={unit}&q=&lat={lat}&lon={lon}"
    response = requests.get(query_url).json()

    if 'sunrise' not in response['sys']:
        sunrise_value = 0
    else:
        sunrise_value = response['sys']['sunrise']
    sunrise_list.append(sunrise_value)
    
    if 'sunset' not in response['sys']:
        sunset_value = 0
    else:
        sunset_value = response['sys']['sunset']
    sunset_list.append(sunset_value)
    
    weather_list = response['weather'][0]['description']
```

```python
subset['SUNRISE'] = sunrise_list
subset['SUNSET']  = sunset_list
subset['WEATHER'] = weather_list
subset
```
